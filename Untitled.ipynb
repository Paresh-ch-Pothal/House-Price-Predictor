```python
def convert_total_sqft(x):
    num=x.split("-")
    if len(num) == 2:
        return float((float(num[0])+float(num[1]))//2)
    try:
        return float(x)
    except:
        return None
```

# Outlier Detection

```python
df[df["total_sqft"]/df.bhk < 300]
```

```python
# Removing the price per square foot outliers
# mean-st and mean+st in normal distribution
def remove_pps_outliers(df1):
    df_out = pd.DataFrame()
    for key, subdf in df1.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df], ignore_index=True)
    return df_out
```

```python
df1=remove_pps_outliers(df)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge, Lasso, ElasticNet
def objective(trial):
    # n_estimators=trial.suggest_int('n_estimators',50,200)
    # max_depth=trial.suggest_int('max_depth',3,20)
    # model=RandomForestRegressor(n_estimators=n_estimators,max_depth=max_depth,
    # random_state=42)
    # param_grid = {
    #     "n_estimators": trial.suggest_int("n_estimators", 100, 1000, step=50),
    #     "max_depth": trial.suggest_int("max_depth", 3, 15),
    #     "learning_rate": trial.suggest_loguniform("learning_rate", 0.01, 0.3),
    #     "subsample": trial.suggest_uniform("subsample", 0.5, 1.0),
    #     "colsample_bytree": trial.suggest_uniform("colsample_bytree", 0.5, 1.0),
    #     "reg_alpha": trial.suggest_loguniform("reg_alpha", 1e-8, 1.0),
    #     "reg_lambda": trial.suggest_loguniform("reg_lambda", 1e-8, 1.0),
    #     "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
    #     "gamma": trial.suggest_loguniform("gamma", 1e-8, 1.0),
    # }
    # model = XGBRegressor(**param_grid, random_state=10)
    model_type = trial.suggest_categorical("model_type", ["Ridge", "Lasso", "ElasticNet"])
    
    # Define hyperparameter search space
    if model_type == "Ridge":
        params = {
            "alpha": trial.suggest_loguniform("alpha", 1e-5, 10),
        }
        model = Ridge(**params)
    elif model_type == "Lasso":
        params = {
            "alpha": trial.suggest_loguniform("alpha", 1e-5, 10),
        }
        model = Lasso(**params)
    elif model_type == "ElasticNet":
        params = {
            "alpha": trial.suggest_loguniform("alpha", 1e-5, 10),
            "l1_ratio": trial.suggest_uniform("l1_ratio", 0.0, 1.0),
        }
        model = ElasticNet(**params)
    pipe=make_pipeline(column_trans,model)
    pipe.fit(xtrain,ytrain)
    ypred=pipe.predict(xtest)
    score=r2_score(ytest,ypred)
    return score
```

```python
import optuna
study=optuna.create_study(direction="maximize",sampler=optuna.samplers.TPESampler())
study.optimize(objective,n_trials=50)
```
